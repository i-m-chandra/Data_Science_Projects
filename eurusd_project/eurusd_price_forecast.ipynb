# Plotting the line chart

```python
sns.lineplot(data=sa_data.head())
```

```python
plt.figure(figsize=(20,10))
plt.title("eurusd")
sns.lineplot(data=sa_data.head(15))
```

```python
plt.figure(figsize=(20,10))
plt.title("eurusd")
sns.lineplot(data=sa_data['Open'].tail(100), label="Open")
sns.lineplot(data=sa_data['High'].tail(100), label="High")
sns.lineplot(data=sa_data['Low'].tail(100), label="Low")
sns.lineplot(data=sa_data['Close'].tail(100), label="Close")
plt.xlabel("Time")
```

# Plotting the volume as bar chart

```python
plt.figure(figsize=(20,10))
plt.title("Volume vs time EURUSD")
sns.barplot(x=sa_data.head(1000).index,y=sa_data.head(1000)['Volume'])
plt.ylabel("Volume")
```

# Plotting Heatmap


```python
plt.figure(figsize=(20,10))
plt.title("EURUSD")
sns.heatmap(data=sa_data,annot=True)
plt.xlabel("Price")
```
